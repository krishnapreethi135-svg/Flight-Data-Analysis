```python
df.loc[:,"Route"]=df["Route"].fillna(df["Route"].mode()[0])
df.loc[:,"Total_Stops"]=df["Total_Stops"].fillna(df["Total_Stops"].mode()[0])
```

```python
plt.figure()
df["Airline"].value_counts().plot(kind="bar",color="purple")
plt.xlabel("Airline")
plt.ylabel("Number of flights")
plt.title("Flights by Airline")
plt.show()
```

```python
print("The above Bar chart graph shows that 'Jet Airways' has the higest number of flights")
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x="Airline",y="Price",data=df,estimator=np.mean,errorbar=("ci",95),color="purple")
plt.xticks(rotation=45)
plt.xlabel("Airline")
plt.ylabel("Average Ticket Price")
plt.title("Average Flight Price by Airline")
plt.show()
```

```python
print("The bar chart shows the average flight ticket prices for different airlines.The airlines with lowest average ticket prices making them more affordable option for travelers")
```

```python
sns.boxplot(x="Total_Stops",y="Price",data=df,color="purple")
```

```python
print("The box plot shows the distribution of flight ticket prices based on number of stops.It is observed that non-stop flights have lower and more consistent prices,with fewer extream values.Flights with 1 or 2 stops show a wider price range,indicating higher variability in ticket prices.As the number of stops increases beyond two,the number of available flights decreases,and prices tend to stabilize,although such flights are less common")
```

```python
stop_counts=df.groupby("Total_Stops")["Airline"].nunique()
stop_counts.plot(kind="bar",color="purple")
plt.xlabel("Number of Stops")
plt.ylabel("Number of Airlines")
plt.title("Airlines by Number of Stops")
plt.show()
```

```python
print("The bar chart shows  the number of airlines offering flight based on the number of stops.Most airlines offer flights with 1 stop,making it the most common option.Non-stop flights are also widely available.As the number of stops increases,fewer airlines provide such flights ,with very few offering 3 or 4 stop routes")
```

```python
Airline_Revenue=df.groupby("Airline")["Price"].sum().reset_index()
plt.figure(figsize=(10,10))
plt.pie(Airline_Revenue["Price"],labels=Airline_Revenue["Airline"],autopct="%1.1f%%",startangle=140)
plt.title("Total Revenue of Airlines")
plt.legend(Airline_Revenue["Airline"],title="Airlines",loc="center left",bbox_to_anchor=)
plt.show()
```

```python
print("The pie chart shows that Jet  Airways has the highest number of flights,followed by Air  India and Jet Airways Business.Indigo also has a significant share,while other airlines like spicejet,GoAir,vistara and Air Asia contribute smaller portions.")
```

```python
destination_counts=df["Destination"].value_counts().reset_index()
destination_counts.columns=["Destination","Flight_Count"]
plt.figure(figsize=(10,6))
sns.lineplot(x="Destination",y="Flight_Count",data=destination_counts,marker="o",color="purple")
plt.xticks(rotation=45)
plt.title("Number Of Flights By Destination")
plt.xlabel("Destination")
plt.ylabel("Number of Flights")
plt.show()
```

```python
print("The line chart shows that Cochin has the highest number of flights,followed by Bangalore.Delhi and New Delhi have a moderate number of flights,while Hyderabad and Kolkata have the lowest")
```

```python
source_counts=df["Source"].value_counts().reset_index()
source_counts.columns=["Source","Flight_Count"]
plt.figure(figsize=(10,6))
sns.barplot(x="Source",y="Flight_Count",data=source_counts,color="purple")
plt.xticks(rotation=45)
plt.title("Number Of Flights per Source City")
plt.xlabel("Source")
plt.ylabel("Number of Flights")
plt.show()
```

```python
print("Delhi is the busiest source city in the dataset,with the highest number of departing flights,   indicating its role as a major aviation hub.")
```
